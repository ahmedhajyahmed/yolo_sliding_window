```python
import os
#path = './annotations'
# path= 'C:/Users/Ahmed/Desktop/croped_db/all_800/annotations'
# path = 'ground_truth'
path = 'C:/Users/Ahmed/Desktop/croped_db/Haj_cropped_clean/annotations'
all_annotations = os.listdir(path)

total_personnes = 0
total_vehicule = 0
personnes_only_images = []
vehicule_only_images = []
for annotation  in all_annotations:
    images_personnes = 0
    images_vehicules = 0
    with open(os.path.join(path, annotation), 'r') as file:
        objects = file.readlines()
        for obj in objects:
            if obj[0] == '0':
                images_personnes +=1
            else:
                images_vehicules +=1
        if images_personnes ==0 and images_vehicules !=0:
            vehicule_only_images.append({"image":annotation, "nb_vehicule":images_vehicules})
        if images_vehicules ==0 and images_personnes!=0:
            personnes_only_images.append({"image":annotation, "nb_personne":images_personnes})
        total_personnes += images_personnes
        total_vehicule += images_vehicules
print(len(all_annotations))      
```

```python
total_objects = total_personnes + total_vehicule
print("total objects = %d" %(total_objects))
print("total personnes = %d => %f percent" %(total_personnes ,100*total_personnes / total_objects))
print("total vehicule = %d => %f percent" %(total_vehicule ,100*total_vehicule / total_objects))
print("there are %d images with only personnes " %(len(personnes_only_images)))
# personnes_only_images
print("there are %d images with only vehicules " %(len(vehicule_only_images)))
# vehicule_only_images
```

```python
many_vehicule_images = []
mvi_obj_nb=0
few_vehicule_images= []
fvi_obj_nb=0
for image in vehicule_only_images:
    if int(image['nb_vehicule']) < 10:
        few_vehicule_images.append(image)
        fvi_obj_nb+= int(image['nb_vehicule'])
        
    else:
        many_vehicule_images.append(image)
        mvi_obj_nb+=int(image['nb_vehicule'])

print(len(few_vehicule_images), fvi_obj_nb)
few_vehicule_images
```
